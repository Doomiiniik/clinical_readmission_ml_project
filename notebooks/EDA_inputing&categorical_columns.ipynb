```python

df = preprocess.clean_missing_values(df)



missing_summary = pd.DataFrame({
    'Count of missing values': df.isna().sum(),
    '(%)': df.isna().mean() * 100
})


missing_summary = missing_summary.sort_values(by='Count of missing values', ascending=False)


missing_summary
```

```python


df = preprocess.filter_eligible_population(df)
df = preprocess.binarize_target(df)
df = preprocess.engineer_features(df)
df = preprocess.encode_categorical_features(df)


train_df, test_df = data_loader.split_data_clinically(
    df,
    target_col="readmitted",
    group_col="patient_nbr"
)
categorical_columns = [
    'max_glu_serum', 'A1Cresult', 'metformin', 'repaglinide', 'nateglinide',
    'chlorpropamide', 'glimepiride', 'acetohexamide', 'glipizide', 'glyburide',
    'tolbutamide', 'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol',
    'troglitazone', 'tolazamide', 'examide', 'citoglipton', 'insulin',
    'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone',
    'metformin-rosiglitazone', 'metformin-pioglitazone', 'change', 'diabetesMed'
]


data_summary = []
for col in categorical_columns:
    unique_values = df[col].dropna().unique()
    data_summary.append({
        'Kolumna': col,
        'Liczba unikalnych wartości': len(unique_values),
        'Wartości': ', '.join(map(str, unique_values))
    })

# Konwertujemy na DataFrame i wyświetlamy
summary_df = pd.DataFrame(data_summary)
pd.set_option('display.max_colwidth', None)  # pokazuje pełne wartości w kolumnie
summary_df

```
